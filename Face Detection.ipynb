```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
face_cascade = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
img = cv2.imread('/content/human pic.webp')
cv2_imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
cv2_imshow(gray)
```

```python
faces = face_cascade.detectMultiScale(gray)
print(faces)
```

```python
for (w, x, y, z) in faces:
    cv2.rectangle(img, (w, x), (w+y, x+z), (255, 100, 0), 2)
cv2_imshow(img)
```

```python
eye_cascade = cv2.CascadeClassifier('/content/haarcascade_eye.xml')
img = cv2.imread('/content/human pic.webp')
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
eyes = eye_cascade.detectMultiScale(gray)
print(eyes)
```

```python
for (w, x, y, z) in eyes:
    cv2.rectangle(img, (w, x), (w+y, x+z), (100, 10, 250), 2)
cv2_imshow(img)
```

```python
def detect(img_path):
  face_cascade = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
  img = cv2.imread(img_path)
  gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
  faces = face_cascade.detectMultiScale(gray)
  for (w, x, y, z) in faces:
    cv2.rectangle(img, (w, x), (w+y, x+z), (200, 100, 0), 2)
  cv2_imshow(img)
```

```python
detect('/content/pictures.jpg')
```
